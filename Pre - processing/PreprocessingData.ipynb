```python
import pandas as pd
import numpy as np
import keras
import tensorflow as tf
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
btc.iloc[btc.shape[0]//2:]
```

```python
y_test =  btc['Close'].iloc[(btc.shape[0]//2):].values
```

```python
y_test.reshape((-1,1))
```

```python
train_generator = TimeseriesGenerator(btc.iloc[:btc.shape[0]//2], btc['Close'].iloc[:btc.shape[0]//2], length=15, batch_size=20)
```

```python
test_generator = TimeseriesGenerator(btc.iloc[(btc.shape[0]//2):], y_test.reshape((-1,1)), length=15, batch_size = 1)
```

```python
test_generator.targets
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

model = Sequential()
model.add(
    LSTM(10,
        activation='relu',
        input_shape=(15, len(btc.columns)))
)
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

num_epochs = 25
model.fit_generator(train_generator, epochs=20, verbose=1)
```

```python
pred = model.predict(test_generator, verbose=1)
```
