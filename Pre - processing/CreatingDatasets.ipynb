```python
import numpy as np
import pandas as pd
import time

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
pd_btc = pd.read_csv('BTCUSDT_5_min_may.csv', header=None)
```

```python
pd_btc.columns = ['Unix timestamp',
'Open',
'High',
'Low',
'Close',
'Volume',
'Close time',
'Quote asset volume',
'Number of trades',
'Taker buy base asset volume',
'Taker buy quote asset volumne',
'Unsure']
```

```python
pd_btc
```

```python
pd_btc['Unix Sec'] = pd_btc['Unix timestamp']//1000
```

```python
pd_btc
```

```python
pd_btc['Unix Sec']
```

```python
tot_time = []
for t in pd_btc['Unix Sec']:
    aux_time = []
    aux_time = time.gmtime(t)
    tot_time.append(aux_time)
```

```python
time_df = pd.DataFrame(tot_time)
```

```python
time_df.loc[16340:16346]
```

```python
aux = time_df[4][0]
special = []
for i in range(1, len(time_df[4])):
    aux2 = time_df[4][i]
    sub = aux2 - aux
    if sub%5 != 0:
        special.append(i)
        print(i)
    aux = aux2
```

```python
time_df.loc[special[0]-2:special[0]+2]
```

```python
if special != []:
    for i in range(special[0], special[1]):
        time_df[4][i] += 2
        if time_df[4][i] == 60:
            time_df[4][i] = 0
            time_df[3][i] += 1
            if time_df[3][i] == 24:
                time_df[3][i] = 0
                time_df[2][i] += 1
                time_df[6][i] += 1
```

```python
time_df[4].value_counts()
```

```python
time_df
```

```python
time_df = time_df.drop([5, 7, 8], axis = 1)
```

```python
time_df.columns = ['Year', 'Month', 'Day', 'Hour', 'Minute', 'Weekday']
```

```python
for i in time_df.columns:
    pd_btc.insert(len(pd_btc.columns), i, time_df[i])
```

```python
pd_btc = pd_btc.drop('Close time', axis = 1)
```

```python
pd_btc = pd_btc.drop(['Unsure', 'Unix Sec'], axis = 1)
```

```python
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(pd_btc['Weekday'])
print(integer_encoded)
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
```

```python
wd_df = pd.DataFrame(onehot_encoded)
```
